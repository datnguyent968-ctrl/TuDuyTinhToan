Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình. 

```python
a, b, c = map(int,input().split())
p = (a + b + c)/2
if a + b > c and a + c > b and b + c > a:
    if a == b and b == c and c == a:
        print("Tam giac deu", a + b + c, (p*(p -a)*(p - b)*(p - c))**(1/2))
    elif a == b and b != c:
        print("Tam giac can", a + b + c, (p*(p -a)*(p - b)*(p - c))**(1/2))
    elif (a**2 + b**2 == c**2) or (c**2 + b**2 == a**2) or (a**2 + c**2 == b**2):
        print("Tam giac vuong", a + b + c, (p*(p -a)*(p - b)*(p - c))**(1/2))
    elif a !=b and b != c and c != a:
        print("Tam giac thuong", a + b + c, (p*(p -a)*(p - b)*(p - c))**(1/2))
else:
    print("K la tam giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0) 


```python
a = float(input())
b = float(input())
c = float(input())

if a == 0:
    if b == 0:
        if c == 0:
            print("Vo so nghiem")
        else:
            print("Vo nghiem")
    else:
        x = -c / b
        print(x)
else:
    delta = b*b - 4*a*c

    if delta < 0:
        print("Vo nghiem")
    elif delta == 0:
        x = -b / (2*a)
        print(x)
    else:
        sqrt_delta = delta ** 0.5
        x1 = (-b + sqrt_delta) / (2*a)
        x2 = (-b - sqrt_delta) / (2*a)
        print(x1, x2)
```

Bài 8: Giải hệ phương trình tuyến tính: 
ax  by  m 
cx  dy  n 

```python
a, b, c, d, m, n = map(float,input().split())
D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c

if D == 0:
    if Dx == 0 and Dy == 0:
        print("Vo so nghiem")
    else:
        print("Vo nghiem")
else:
    x = Dx / D
    y = Dy / D
    print(x, y)
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a, b, R, x, y = map(float,input().split())

trai = (x - a) * (x - a) + (y - b) * (y - b)
phai = R * R

if trai == phai:
    print("Thuoc duong tron")
else:
    print("Khong thuoc duong tron")
```
