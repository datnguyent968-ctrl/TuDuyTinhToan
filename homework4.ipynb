W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 


```python
n = int(input())

while n > 1 and n % 2 == 0:
    n = n // 2

if n == 1:
    print("La dang 2^k")
else:
    print("Khong phai dang 2^k")
```

```python
n = int(input())

tong = 0
while n > 0:
    tong = tong + n % 10
    n = n // 10

print(tong)
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())

chan = 0
le = 0

while n > 0:
    cs = n % 10
    if cs % 2 == 0:
        chan = chan + 1
    else:
        le = le + 1
    n = n // 10

print(chan, le)
```

```python
n = int(input())

S = 0
k = 0

while S + (k + 1) < n:
    k = k + 1
    S = S + k

print(k)
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
x = int(input())

max_val = x
min_val = x

while True:
    x = int(input())
    if x == -1:
        break

    if x > max_val:
        max_val = x

    if x < min_val:
        min_val = x

print(max_val, min_val)
```

W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
A = int(input())

f1 = 1
f2 = 1

if A == 1:
    print(1)
else:
    while f1 + f2 <= A:
        f3 = f1 + f2
        f1 = f2
        f2 = f3
    print(f2)
```

W4A27
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), l

```python
s = input()

dem = 0
i = 0

while True:
    try:
        s[i]
        if i == 0 or s[i - 1] == ' ':
            dem = dem + 1
        i = i + 1
    except:
        break

print(dem)
```
